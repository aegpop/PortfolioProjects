```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (replace this with your actual DataFrame)
# Assuming your DataFrame is named df
# Replace this with your actual DataFrame
# df = ...

# Define the revenue bins with a range of $10,000
revenue_bins = range(0, int(df['Revenue'].max()) + 10000, 10000)

# Group by 'Revenue Range' and sum the revenue in each bin
revenue_per_range = df.groupby(pd.cut(df['Revenue'], bins=revenue_bins, right=False))['Revenue'].sum()

# Calculate the total revenue
total_revenue = df['Revenue'].sum()

# Plot the bar chart
plt.figure(figsize=(10, 6))
revenue_per_range.plot(kind='bar', color='skyblue', label='Revenue per Range')
plt.axhline(y=total_revenue, color='red', linestyle='--', label='Total Revenue')
plt.xlabel('Revenue Range')
plt.ylabel('Revenue')
plt.title('Revenue Comparison by Range')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
